#Handling Missing Values

```python
# check missing values

data.isnull().sum()

# choose the method of inputting missing value
# E,g Numerical ->  mean, median or KNNInputer
# Discreate and Continous:Life Expectancy and Adult Mortality(Missing Values are filled by the Mean)
# Categorical variables: we fill by the Mode.

data['Life expectancy '].mean()

# a) Filling the Missing Records -> fill_na()
# We normally use df[col] = df[col].fillna(df[col].mean()) #for mean imputation
data['Life expectancy '] = data['Life expectancy '].fillna(data['Life expectancy '].mean())
data['Adult Mortality'] = data['Adult Mortality'].fillna(data['Adult Mortality'].mean())
data['Alcohol'] = data['Alcohol'].fillna(data['Alcohol'].mean())
data['Hepatitis B'] = data['Hepatitis B'].fillna(data['Hepatitis B'].mean())
data[' BMI '] = data[' BMI '].fillna(data[' BMI '].mean())
data['Polio'] = data['Polio'].fillna(data['Polio'].mean())
data['Total expenditure'] = data['Total expenditure'].fillna(data['Total expenditure'].mean())
data['GDP'] = data['GDP'].fillna(data['GDP'].mean())
data['Population'] = data['Population'].fillna(data['Population'].mean())
data['Income composition of resources'] = data['Income composition of resources'].fillna(data['Income composition of resources'].mean())

data.isnull().sum()

# b)Dropping the Empties -> drop_na(commit)
data.dropna(inplace=True)

# c) Imputation Method -> Machine Learning


data.isnull().sum() #check to see if there are more empty rows
```

#Detecting Outliers

```python
# OUTLIERS DETECTION USING BOXPLOTS
# OUTLIERS:
# Outliers are extreme values that deviate significantly from other observations.
# They can be caused by errors in data entry or genuine variations in the dataset.

# Age column -> 200
# Life expectancy column -> 150
# Life expenctancy column -> 1


# DETECTING OUTLIERS USING BOX-PLOTS
# E,g Life Expenctancy
sns.boxplot(data = data, x = 'Polio')
plt.title('Outliers in Polio')
plt.show()


# DETECTING OUTLIERS FOR ALL THE NUMERIC COLUMNS
#print(data.select_dtypes(include='number').columns)

for i in data.select_dtypes(include='number').columns:
    sns.boxplot(data = data, x = i)
    plt.title(f'Outliers in {i}')
    plt.show()
```

#Removing Outliers

```python
# Detecting Outliers -> Box Plot
# Remove the Outliers -> Winsorization –
#  Instead of removing outliers, extreme values are replaced with upper or lower percentile values.

def remove_outliers(col):
    q1, q3 = np.percentile(col, [0.25, 0.75])
    iqr = q3-q1
    lw = q1-(1.5*iqr)
    uw = q3+(1.5*iqr)
    return lw, uw

remove_outliers(data['Schooling'])
# Remove outliers
for i in data.select_dtypes(include='number').columns:
  lw, uw = remove_outliers(data[i])
  data[i] = np.where(data[i]<lw, lw, data[i])
  data[i] = np.where(data[i]>uw, uw, data[i])
  sns.boxplot(data=data,x=i)
  plt.show()
```
